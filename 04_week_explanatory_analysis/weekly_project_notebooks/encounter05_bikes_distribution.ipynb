## Task Description

Create a notebook that is solving following questions:

   1. Filter bike data for one day
   2. Using `sns.displot()` plot the distribution of the count during that day (Try setting the parameter `kde=True`).
   3. Filter bike data for the year 2011
   4. Again using `sns.displot()` plot the distribution of the temperature of the entire year

```python
#import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# reading 'bikes' dataset from csv-file
bikes = pd.read_csv('../data/bikes_with_bins.csv', index_col='datetime', parse_dates=True)
bikes.head()
```

### 1. Filter bike data for one day

```python
bikes_20110621 = bikes.loc['2011-06-21']
bikes_20110621.sample(5)
```

```python
# bikes_20110621.info() 
# DatetimeIndex: 24 entries, 2011-06-21 00:00:00 to 2011-06-21 23:00:00
# Correct!
```

### 2. Using sns.displot() plot the distribution of the count during that day (Try setting the parameter kde=True).

```python
sns.displot(bikes_20110621, x='count', kde = True, stat="probability");
#sns.displot(bikes_20110621, x='count');
```

```python
# adding hue = 'part_of_day'
sns.displot(data=bikes_20110621, x='count', hue='part_of_day', kind = 'kde');
```

### 3. Filter bike data for the year 2011

```python
mask_2011 = bikes['year'] == 2011
bikes_2011 = bikes[mask_2011]
bikes_2011.sample()
```

```python
# bikes_2011.info()
# DatetimeIndex: 8645 entries, 2011-01-01 00:00:00 to 2011-12-31 23:00:00
```

### 4. Again using sns.displot() plot the distribution of the temperature of the entire year

```python
sns.displot(bikes_2011, x='temp', kde = True, stat="probability");
```

```python
sns.displot(data=bikes_2011, x='temp', hue='part_of_day', kind = 'kde');
```
