```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 1

Open empty jupyter notebook and read in the dataframe that contains your gapminder data

```python
gapminder = pd.read_csv('../data/gapminder_total.csv')
gapminder
```

```python
# converting 'year' and 'population' to 'Int64'
# converting 'year' and 'population' datatype to 'Int64'
gapminder['year'] = gapminder['year'].astype('Int64')
#gapminder['population'] = gapminder['population'].astype('Int64')

gapminder.info()
```

### Step 2

Create subset of data for one year

```python
# filtering out 2015 data using boolean mask
gapminder_2015 = gapminder[gapminder['year'] == 2015]
gapminder_2015
```

### Step 3

Plot the life expectancy vs fertility rate

```python
sns.scatterplot(x='life expectancy', y='fertility', data=gapminder_2015);
```

### Step 5

Using matplotlib add a title and labels. Increase the plot size to make it more readable.

```python
plt.figure(figsize=(8, 4))

sns.scatterplot(x='life expectancy', y='fertility', data=gapminder_2015);

plt.title('Life Expectancy vs Fertility Rate in 2015')
plt.ylabel('Fertility')
plt.xlabel('Life Expectancy')
```

### Step 6

Weight the size of the scatter points to the population of each country. This can be done using the `size` parameter in the `sns.scatterplot()` function.

**Tip:** the code above is correct but the magnitude of the population is astronomically larger than that of the life expectancy and fertility rate. Use some math to fix this.

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(x='life expectancy', y='fertility', 
                data=gapminder_2015,
                size='population',
                #s=(gapminder_2015['population']/1_000_000),
                sizes=(20, 400),
                hue='continent'
               );

plt.title('Life Expectancy vs Fertility Rate in 2015')
plt.ylabel('Fertility')
plt.xlabel('Life Expectancy')
```

### Step 7

Setting the axes to display the origin also gives us a more complete picture of what is going on.

```python
# plt.axis([0, 85, 0, 9.5])

plt.figure(figsize=(8, 6))

sns.scatterplot(x='life expectancy', y='fertility', 
                data=gapminder_2015,
                size='population',
                sizes=(20, 400),
                hue='continent'
               );

plt.axis([0, 85, 0, 9.5])
plt.title('Life Expectancy vs Fertility Rate in 2015')
plt.ylabel('Fertility')
plt.xlabel('Life Expectancy')

# saving the image as png
plt.savefig('../data/plot_fig_2015.png')
```
