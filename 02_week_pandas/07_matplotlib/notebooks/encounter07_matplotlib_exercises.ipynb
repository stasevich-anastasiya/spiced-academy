## Task 2: Plot a quadratic function

   * create a list of x values ranging from -10 to 10
   * create an empty list for the y values
   * loop over the values of x and calculate y=x^2^
   * append each value of y to the list
   * make a lineplot of the data
   * add a legend, title and axis labels to the plot

```python
# create a list of x values ranging from -10 to 10
x = [x for x in range(-10,11)]
x
```

```python
# create an empty list for the y values
y =[]
y
```

```python
# loop over the values of x and calculate y=x2y=x2
# append each value of y to the list
for i in x:
    y.append(i**2)
    
y
```

```python
# make a lineplot of the data
# add a legend, title and axis labels to the plot

# lineplot
plt.plot(x, y, label='Task 2 Lineplot');

# legend
plt.legend();

# title
plt.title('Some Random Data');

# axis labels
plt.xlabel('x values');
plt.ylabel('y values');
```

```python
# create a list of weekdays:
x = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su']
```

```python
# For each day of the week write down the number of times you called someone and use those figures to make a list called y
y = [5, 0, 2, 1, 10, 2, 0]
```

```python
# Visualize the data with a bar plot
# Change the color of the bars, add a tile

# plt.figure(figsize=(8, 3)); -> Commented! Reason: the saved image is cut and has n 'x axis' description !!

plt.bar(x, y, color=['darkorange', 'white', 'orange', 'green', 'indianred','orange', 'white'], width=0.8);

# title
plt.title('My Week on the Phone');

# axis labels
plt.xlabel('Day of Week');
plt.ylabel('Number of Calls');

# Save the plot as a .png file
plt.savefig('../images/encounter7_task3.png')
plt.savefig('../images/encounter7_task3.pdf')
```
