```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
name = 'Lebron'

df_name = pd.read_csv(f'../data/{name}_all_years.csv')
df_name
```

```python
# checking if there are a few recs for any particular year (year dups)
df_name.nunique()
```

As far as there are no years with (male, female) pair of names, no additional DataFrame filtering is required. 

```python
# Create a line plot with the years and the (relative) frequencies of a specific name

x_vals = df_name['year']
y_vals = df_name['frequency']

# plotting starts
plt.style.use('ggplot')
plt.figure(figsize=(10, 5))

plt.plot(x_vals, y_vals, label='name frequency per year', color='darkorange', linewidth=5)

# title
plt.title('The (relative) frequencies of "Lebron" name per year from 1880 to 2020')

# label names
plt.xlabel('Years')
plt.ylabel('Name Frequency')

# save plot at given path
#plt.savefig('../images/some_random_plot.pdf')
```
