## Task Description

Create a plot of the summed up frequency (total births) per year from 1880 to 2020.

### Step 1 - Create the data

    1. Make a empty dictionary that maps the column names year and total_births to empty lists:

> births_dict = {'year':[], 'total_births':[]}

    2. Loop over the years from 1880 to 2020
    3. Use your parse_dataset(year) function to parse each year
    4. Use the pandas.sum() method to calculate the sum of the frequencies (births)
    5. Append the result to the dictionary using birth_dict['total_births'].append()
    6. Also append the year to birth_dict['year']
    7. Create a yearly_births DataFrame by passing births_dictto pd.DataFrame()

This should result in a DataFrame that has the year as one column and the total_birthsin that year in another column.


```python
import pandas as pd
import matplotlib.pyplot as plt
import os  # for checking if the file with constructed puth exists
```

```python
# my parse_dataset(year) function

def parse_dataset(year):
    '''Function accepts year, reads appropriate data file and returns DataFrame with header'''
    
    # Check if entered YEAR is at least an INT number; if not, return a text message
    if type(year) is int:
        path = f'../data/yob{year}.txt'
        if os.path.isfile(path):
            dataset = pd.read_csv(path, names=col_list)
            return dataset
        else:
            return print(f'There is no file with entered YEAR = \'{year}\'!')
    
    else:
        return print(f'Entered YEAR = \'{year}\' is not an INT value!')
```

```python
# Make a empty dictionary that maps the column names year and total_births to empty lists:
birth_dict = {'year':[], 'total_births':[]}

# list of columns in DataFrame
col_list=['name', 'gender', 'frequency']
```

```python
# Loop over the years from 1880 to 2020
# Use your parse_dataset(year) function to parse each year
# Use the pandas.sum() method to calculate the sum of the frequencies (births)
# Append the result to the dictionary using birth_dict['total_births'].append()
# Also append the year to birth_dict['year']

for year in range(1880,2021):
    # parsing current year dataset
    df_temp = parse_dataset(year)
    
    #counting total_births
    total_births = df_temp['frequency'].sum()
     
    # adding 'year' and 'total' to 'birth_dict' dictionary
    birth_dict['total_births'].append(total_births)
    birth_dict['year'].append(year)

# birth_dict
```

```python
# Create a yearly_births DataFrame by passing births_dictto pd.DataFrame()

yearly_births = pd.DataFrame(birth_dict)
yearly_births
```

### Step 2 - Plotting

   * use pandas to make a quick plot of your results
   * use matplotlib to make a more detailed plot but adding a legend, title and and axis labels
   * try other plot types

```python
# use pandas to make a quick plot of your results
yearly_births.plot(x='year', y='total_births');
```

```python
# use matplotlib to make a more detailed plot but adding a legend, title and and axis labels

# defining values for x and y axis
x_vals = yearly_births['year']
y_vals = yearly_births['total_births']

plt.style.use('ggplot')
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y_vals, label='births amount', linewidth=3)

# title
plt.title('Summed up total births per year from 1880 to 2020 in USA')

# label names
plt.xlabel('Years')
plt.ylabel('Totals')

# creating a Legend: uses 'label' value from plt.plot() line
plt.legend();
```

```python
# bar

# defining values for x and y axis
x_vals = yearly_births['year']
y_vals = yearly_births['total_births']

plt.style.use('ggplot')
plt.figure(figsize=(10, 5))
plt.bar(x_vals, y_vals, label='births amount', linewidth=0.2)

# title
plt.title('Summed up total births per year from 1880 to 2020 in USA')

# label names
plt.xlabel('Years')
plt.ylabel('Totals')

# creating a Legend: uses 'label' value from plt.plot() line
plt.legend();
```

```python
# scatter plot

# defining values for x and y axis
x_vals = yearly_births['year']
y_vals = yearly_births['total_births']

plt.style.use('ggplot')
plt.figure(figsize=(10, 5))
plt.scatter(x_vals, y_vals, label='births amount', linewidth=0.1)

# title
plt.title('Summed up total births per year from 1880 to 2020 in USA')

# label names
plt.xlabel('Years')
plt.ylabel('Totals')

# creating a Legend: uses 'label' value from plt.plot() line
plt.legend();
```
